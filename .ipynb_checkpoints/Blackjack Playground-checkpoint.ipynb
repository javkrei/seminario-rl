Blackjack es un juego de cartas donde el objetivo es sumar lo más cercano a 21 sin pasarse. En este caso la estrategia del crupier es fija. 

La J, reina y el rey valen 10. El as puede valer 11 o 1, y se denomina 'usable' si puede tomar el valor 11 en la mano sin pasarse de 21. Las cartas pueden tomarse con reemplazo o de un mazo infinito. 

El juego comienza con cada jugador con una carta boca abajo y una boca arriba. El jugador puede pedir cartas adicionales (un 'hit') hasta que decide plantarse ('stick') o pierde ('bust'). Después de que el jugador se planta, el crupier saca cartas hasta que su mano es mayor o igual a 17. Si el crupier obiene más de 21 el jugador gana. Si ni el jugador ni el crupier obtienen más de 21 entonces gana quien esté más cerca de 21, o empatan si el valor es igual, las recompensas para el jugador son las siguientes: por ganar +1, por perder -1, por empatar 0. 

La observación/estado es una tupla de 3 elementos: la suma actual del jugador, la carta que está mostrando el crupier, y si el jugador tiene un as usable o no. Este environment corresponde a la versión del problema del blackjack descripta en el ejemplo 5.1 de Reinforcement Learning: An Introduction by Sutton and Barto (1998).

```python
env = BlackjackEnv()
```

```python
def print_observation(observation):
    score, dealer_score, usable_ace = observation
    print("Mano del jugador: {} (As usable: {}), Mano del crupier: {}".format(
          score,["No", "Sí"][usable_ace] , dealer_score))

def strategy(observation):
    score, dealer_score, usable_ace = observation
    # Stick (action 0) if the score is > 20, hit (action 1) otherwise
    return 0 if score >= 20 else 1

for i_episode in range(20):
    observation = env.reset()
    for t in range(100):
        print_observation(observation)
        action = strategy(observation)
        print("Tomando acción: {}".format( ["Plantarse", "Pedir carta"][action]))
        observation, reward, done, _ = env.step(action)
        if done:
            print_observation(observation)
            print("Fin del juego. Recompensa: {}\n".format(float(reward)))
            break
```
